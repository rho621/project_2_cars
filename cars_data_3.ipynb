```python
import pandas as pd
```

```python
cars_csv = "cars_database.csv"
```

```python
cars_data_df = pd.read_csv(cars_csv)
cars_data_df = cars_data_df.apply(lambda x: x.astype(str).str.upper())
```

```python
new_car_data = cars_data_df[['Make','Model','Origin','MSRP','Weight','MPG_Highway' ]]
new_car_data.head()
```

```python
#Modify model, try to get only the model name. 
model = new_car_data["Model"]

modified_model=[]


#split string based on spaces and append the first value to a new "modifie_model" list. Make everything lowercase
for name in model:
    name = name.upper()
    split = name.split()
    modified_model.append(split[0])
    
print(modified_model)
    
modified_model_df = pd.DataFrame(modified_model)
```

```python
#modified_model_df.head()
#new_car_data2 = new_car_data.insert("Model",modified_model)
new_car_data["Model"]= modified_model
print(new_car_data.head())
```

```python
first_list = cars_data_df['Make'].unique().tolist()
second_list = new_car_data['Model'].unique().tolist()
```

```python
json_file = "car_database.json"
json_cars_df = pd.read_json(json_file, lines=True)
json_cars_df.head()
```

```python
new_jsoncars_df = json_cars_df[['Brand', 'Model',"Acceleration 0 - 100 km/h"]].copy()
new_jsoncars_df = new_jsoncars_df.apply(lambda x: x.astype(str).str.upper())
```

```python
cars_df = new_jsoncars_df.rename(columns={"Brand": "Make","Acceleration 0 - 100 km/h": "Acceleration (0 - 60 mpg)"}) 
cars_clean_df = cars_df.dropna()
cars_clean_df.head()
```

```python
cars_clean_df['Make'].unique()
df = cars_clean_df[cars_clean_df.Make.isin(first_list)]
#df = cars_clean_df[cars_clean_df.Make.isin(second_list)]
df
```

```python
no_dup_cars = df.drop_duplicates(subset=['Make', 'Model'])
no_dup_cars = no_dup_cars[no_dup_cars.Model.isin(second_list)]
no_dup_cars
```
