```python
import pandas as pd
```

```python
cars_csv = "cars_database.csv"
```

```python
cars_data_df = pd.read_csv(cars_csv)
cars_data_df = cars_data_df.apply(lambda x: x.astype(str).str.lower()
```

```python
first_list = cars_data_df['Make'].unique().tolist()
```

```python
json_file = "car_database.json"
json_cars_df = pd.read_json(json_file, lines=True)
json_cars_df.head()
```

```python
new_jsoncars_df = json_cars_df[['Brand', 'Model',"Acceleration 0 - 100 km/h"]].copy()
new_jsoncars_df = new_jsoncars_df.apply(lambda x: x.astype(str).str.lower())
```

```python
cars_df = new_jsoncars_df.rename(columns={"Brand": "Make","Acceleration 0 - 100 km/h": "Acceleration"}) 
cars_clean_df = cars_df.dropna()
cars_clean_df.head()
```

```python
cars_clean_df['Make'].unique()
df = cars_clean_df[cars_clean_df.Make.isin(first_list)]
df
```

```python
no_dup_cars = df.drop_duplicates(subset=['Make', 'Model'])
no_dup_cars
```

```python
car_list = no_dup_cars['Make'].unique()
car_list
```
