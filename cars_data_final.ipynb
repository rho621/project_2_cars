```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
cars_csv = "cars_database.csv"
```

```python
#converting to csv to data frame and making everything into uppercase
cars_data_df = pd.read_csv(cars_csv)
cars_data_df = cars_data_df.apply(lambda x: x.astype(str).str.upper())
```

```python
#only adding data we want to look at 
new_car_data = cars_data_df[['Make','Model','Origin','MSRP','Weight','MPG_Highway' ]]
new_car_data.head()
```

```python
#Modify model, try to get only the model name. 
model = new_car_data["Model"]

modified_model=[]


#split string based on spaces and append the first value to a new "modifie_model" list. Make everything lowercase
for name in model:
    name = name.upper()
    split = name.split()
    modified_model.append(split[0])
    
print(modified_model)
    
modified_model_df = pd.DataFrame(modified_model)
```

```python
#modified_model_df.head()
#new_car_data2 = new_car_data.insert("Model",modified_model)
new_car_data["Model"]= modified_model
final_csv_df = new_car_data.drop_duplicates(subset=['Model'])
final_csv_df["PK"] = final_csv_df["Make"] + final_csv_df["Model"]
final_csv_df
```

```python
#converting "make" column and "model" column into a list
make_list = final_csv_df['Make'].unique().tolist()
model_list = final_csv_df['Model'].unique().tolist()
```

```python
#converting the json file into a dataframe
json_file = "car_database.json"
json_cars_df = pd.read_json(json_file, lines=True)
json_cars_df.head()
```

```python
#only copying the data for columns that we want to view
new_jsoncars_df = json_cars_df[['Brand', 'Model',"Acceleration 0 - 100 km/h", "Maximum speed"]].copy()
#new_jsoncars_df = new_jsoncars_df.apply(lambda x: x.astype(str).str.upper())
```

```python
new_jsoncars_df["Brand"] = new_jsoncars_df["Brand"].str.upper() 
new_jsoncars_df["Model"] = new_jsoncars_df["Model"].str.upper() 
```

```python
#renaming the columns
cars_df = new_jsoncars_df.rename(columns={"Brand": "Make","Acceleration 0 - 100 km/h": "Acceleration 0-60 mph"}) 
cars_clean_df = cars_df.dropna()
cars_clean_df.head()
```

```python
#dropping any "make" that was not part of the first list
cars_clean_df['Make'].unique()
df = cars_clean_df[cars_clean_df.Make.isin(make_list)]

#removing and duplicate "make" and "model"
df = df.drop_duplicates(subset=['Make', 'Model'])
df
```

```python
#dropping "model" that is not part of the first list
final_json_df = df[df.Model.isin(model_list)]
final_json_df["PK"] = final_json_df["Make"] + final_json_df["Model"]
final_json_df
```

```python
rds_connection_string = "postgres:postgres@localhost:5432/cars_db"
engine = create_engine(f'postgresql://{rds_connection_string}')
```

```python
engine.table_names()
```

```python
final_json_df.to_sql(name='cars_json', con=engine, if_exists='append', index=False)
```

```python
final_csv_df.to_sql(name='cars_csv', con=engine, if_exists='append', index=False)
```
