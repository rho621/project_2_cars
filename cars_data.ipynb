```python
import pandas as pd
```

```python
cars_csv = "cars_database.csv"
```

```python
cars_data_df = pd.read_csv(cars_csv)
cars_data_df.head()
```

```python
json_file = "car_database.json"
json_cars_df = pd.read_json(json_file, lines=True)
json_cars_df.head()
```

```python
new_jsoncars_df = json_cars_df[['Brand', 'Model', 'Acceleration 0 - 100 km/h', "Torque"]].copy()
new_jsoncars_df.head()
```

```python
cars_df = new_jsoncars_df.rename(columns={"Brand": "Make","Acceleration 0 - 100 km/h": "Acceleration"}) 
cars_clean_df = cars_df.dropna()
cars_clean_df
```

```python
no_dup_cars = cars_clean_df.drop_duplicates(subset=['Make', 'Model'], keep=False)
no_dup_cars
```
